# Interactive Visualization of Braunschweiger City Ring

Converting the OpenDRIVE map into a `geopandas` entity has many benefits. One
is visualizing the road network interactively using `folium`.

At first, let's load the road network and convert it to a
`geopandas.GeoDataFrame`. This may take a while.

```python
from xodr2nxgpd.network import RoadNetwork

net = RoadNetwork.from_file().as_geopandas()
net.head()
```

We can now call the `explore()` method of the `GeoDataFrame`. For that purpose,
however, we need to project all geometric entities into another EPSG for proper
visualization. We will use the *32632* reference system.

```python
net.to_crs("EPSG:32632").reset_index().explore(column='road_type', cmap='tab10', tiles="CartoDB positron")
```
